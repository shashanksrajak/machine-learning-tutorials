## Movie Sentiment Analysis from IMDB Movie Database

1. In this Exercise, you are going to classify whether a given movie review is positive or negative.
2. you are going to use Bag of words for pre-processing the text and apply different classification algorithms.
3. Sklearn CountVectorizer has the inbuilt implementations for Bag of Words.

[Download IMDB Dataset First](https://www.kaggle.com/datasets/lakshmi25npathi/imdb-dataset-of-50k-movie-reviews/code)

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

### Load the data

```python
dataset_source = "./IMDB_Dataset.csv"
```

```python
movies_sentiment = pd.read_csv(dataset_source)
movies_sentiment.head(5)
```

```python
movies_sentiment.shape
```

```python
movies_sentiment.info()
```

Update the label for the dataset

```python
movies_sentiment['Category'] = movies_sentiment['sentiment'].apply(lambda y: 1 if y == "positive" else 0)

movies_sentiment.head(5)
```

```python
movies_sentiment['Category'].value_counts()
```

The data is balanced.

Create train and test datasets

```python
X_train, X_test, y_train, y_test = train_test_split(movies_sentiment['review'], movies_sentiment['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape, y_train.shape
```

#### Create a pipeline to vectorize and then classify sentiments

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
clf = Pipeline(
    [
        ("vectorizer", CountVectorizer()),
        ("classify", RandomForestClassifier(criterion="entropy", n_estimators=50))
    ]
)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
# classification report
print(classification_report(y_true=y_test, y_pred=y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = Pipeline([
    ("vectorizer", CountVectorizer()),
    ("classify", KNeighborsClassifier(n_neighbors=10, metric="euclidean"))
])

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
# classification report
print(classification_report(y_true=y_test, y_pred=y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = Pipeline([
    ("vectorizer", CountVectorizer()),
    ("classify", MultinomialNB())
])

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
# classification report
print(classification_report(y_true=y_test, y_pred=y_pred))
```
